## Implement linear regression from scratch using one and multiple variables, and visualise and interpret the resulting models.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)  # true relation: y = 4 + 3x + noise

plt.scatter(X, y, color='blue', alpha=0.6)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Synthetic Data: y = 4 + 3x + noise")
plt.show()
```

```python
X_b = np.c_[np.ones((100, 1)), X]  # add bias term (column of 1s)
theta = np.random.randn(2, 1)      # random initial values for weights
```

```python
# hyperparameters
learning_rate = 0.1
iterations = 1000
m = len(y)
```

```python
cost_history = []

for i in range(iterations):
    gradients = (2/m) * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - learning_rate * gradients
    cost = np.mean((X_b.dot(theta) - y) ** 2) / 2
    cost_history.append(cost)

print(f"Learned Parameters (theta): {theta.ravel()}")
print(f"True Parameters: [4, 3]")
```

```python
plt.scatter(X, y, label="Training Data", alpha=0.6)
plt.plot(X, X_b.dot(theta), color="red", label="Model Prediction")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression (One Variable)")
plt.legend()
plt.show()
```

```python
plt.plot(cost_history)
plt.title("Cost Function Convergence")
plt.xlabel("Iteration")
plt.ylabel("Cost (MSE/2)")
plt.show()
```
